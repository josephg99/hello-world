```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

#-------------------------------------------------data preprocessing------------------------------------------------------------------

# Define the URL of a CSV file containing data.
url = 'https://raw.githubusercontent.com/zhenliangma/Applied-AI-in-Transportation/master/Exercise_7_Neural_networks/Exercise7data.csv'

# Read the CSV data
df = pd.read_csv(url)

# Limit the DataFrame to the first 1000 rows
df = df.iloc[:1000]

# Drop unnecessary columns
df = df.drop(['Arrival_time', 'Stop_id', 'Bus_id', 'Line_id'], axis=1)

# Extract features and target
x = df.drop(['Arrival_delay'], axis=1)
y = df['Arrival_delay']

# Split the train and test data
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

#-------------------------------------------------network construction - loop (layer and dropout)-------------------------------------------

# Define different combinations of layers and dropouts
layer_configs = [(32, 64), (64, 128), (128, 256)] # I here decide the model parameters here
dropout_rates = [0.2, 0.3, 0.4]

results = []

for layers in layer_configs:
    for dropout_rate in dropout_rates:
        print(f"Training model with layers {layers} and dropout {dropout_rate}")

        # Create a Sequential model
        model = Sequential()
        model.add(Dense(layers[0], activation='relu', input_dim=4))
        model.add(Dropout(dropout_rate))
        model.add(Dense(layers[1], activation='relu'))
        model.add(Dropout(dropout_rate))
        model.add(Dense(1))  # Output layer for regression

        # Compile the model
        model.compile(optimizer='adam', loss='mae', metrics=['mae'])

        # Define callbacks
        early_stop = EarlyStopping(monitor='val_mae', patience=5, restore_best_weights=True)
        reduce_lr = ReduceLROnPlateau(monitor='val_mae', factor=0.5, patience=3)
        filepath = "weights.best.keras"
        checkpoint = ModelCheckpoint(filepath, monitor='val_mae', verbose=0, save_best_only=True, mode='min')

        # Train the model
        hist = model.fit(X_train, y_train, validation_split=0.2, epochs=200, batch_size=32,
                         callbacks=[early_stop, reduce_lr, checkpoint], verbose=0)

        # Set the style of the plots using seaborn.
        sns.set()

        # Extract the training and validation Mean Absolute Error (MAE) from the training history.
        err = hist.history['mae']
        val_err = hist.history['val_mae']

        # Define the number of epochs.
        epochs = range(1, len(err) + 1)

        # Plot the Training MAE and Validation MAE over epochs.
        plt.figure(figsize=(8, 4))
        plt.plot(epochs, err, '-', label='Training MAE')
        plt.plot(epochs, val_err, ':', label='Validation MAE')
        plt.title(f'Training and Validation MAE (Layers: {layers}, Dropout: {dropout_rate})')
        plt.xlabel('Epoch')
        plt.ylabel('Mean Absolute Error')
        plt.legend(loc='upper right')
        plt.tight_layout()
        plt.show()

        # Evaluate the model
        y_pred = model.predict(X_test)
        mae = mean_absolute_error(y_test, y_pred)
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)

        # Store the results
        results.append({
            'layers': layers,
            'dropout_rate': dropout_rate,
            'mae': mae,
            'mse': mse,
            'r2': r2
        })

# Display the results
for result in results:
    print(f"Layers: {result['layers']}, Dropout: {result['dropout_rate']}, MAE: {result['mae']:.4f}, MSE: {result['mse']:.4f}, R2: {result['r2']:.4f}")

#-------------------------------------------------visualization------------------------------------------------------------------

# Set the style of the plots
sns.set()

# Extract metrics for plotting
mae_list = [res['mae'] for res in results]
mse_list = [res['mse'] for res in results]
r2_list = [res['r2'] for res in results]
labels = [f"Layers {res['layers']}, Dropout {res['dropout_rate']}" for res in results]

# Plot MAE for different configurations
plt.figure(figsize=(12, 6))
plt.bar(labels, mae_list)
plt.title('Mean Absolute Error (MAE) for Different Configurations')
plt.xticks(rotation=45, ha='right')
plt.ylabel('MAE')
plt.tight_layout()
plt.show()

# Plot MSE for different configurations
plt.figure(figsize=(12, 6))
plt.bar(labels, mse_list)
plt.title('Mean Squared Error (MSE) for Different Configurations')
plt.xticks(rotation=45, ha='right')
plt.ylabel('MSE')
plt.tight_layout()
plt.show()

# Plot R2 for different configurations
plt.figure(figsize=(12, 6))
plt.bar(labels, r2_list)
plt.title('R-squared (R2) for Different Configurations')
plt.xticks(rotation=45, ha='right')
plt.ylabel('R2')
plt.tight_layout()
plt.show()
```
